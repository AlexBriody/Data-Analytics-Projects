- What is the distribution of the `category_name` column?

```python
# Create the subplot
fig, ax = plt.subplots(figsize = (16,9)) 

# Get data
data = kickstarter_data['category_name']

# Plot histogram
plt.hist(data, bins = 100, label = 'Project Categories', color = 'b', alpha = 0.75) 

# Label the histogram
plt.xlabel('Category Name', fontsize = 20, fontname = 'Times New Roman')
plt.ylabel('Frequency', fontname = 'Times New Roman', fontsize = 20)
plt.title('Distribution of Kickstarter Categories', fontsize = 28)
plt.legend(loc = 'upper right', fontsize = 18).get_frame().set_color('#ced6e0')
plt.xticks(rotation = 80)

# Show the plot
plt.tight_layout()
plt.show()
```

- Which country has the most projects?

Answer: The United States (see graph below)

```python
# Get data
data = kickstarter_data['country']

# Create the subplot
fig, ax = plt.subplots(figsize = (16,9)) 

# Plot histogram
plt.hist(data, bins = 100, label = 'Project Countries', color = 'g', alpha = 0.75) 

# Label the histogram
plt.xlabel('Country', fontsize = 20, fontname = 'Times New Roman')
plt.ylabel('Frequency', fontname = 'Times New Roman', fontsize = 20)
plt.title('Distribution of Kickstarter Projects by Country', fontsize = 28)
plt.legend(loc = 'upper right', fontsize = 18).get_frame().set_color('#ced6e0')
plt.xticks(rotation = 90)

# Show the graph
plt.tight_layout()
plt.show()
```

- What are the top 20 Kickstarter projects?

I answered this question by plotting the 'usd_pledged' category, which is the amount of funding pledged in US dollars, against "name." The greater amount of funding, the larger the Kickstarter projects (IDed by 'name'), in my opinon.

```python
# Sort the 'usd_pledged' category, the top 20, with largest first
top_projects = kickstarter_data.sort_values('usd_pledged', ascending=False).head(20)
top_projects = top_projects.iloc[::-1]

# Create the graph - horizontal bar graph
fig, ax = plt.subplots(figsize = (16,9))
plt.barh(top_projects['name'], top_projects['usd_pledged'], color='b', alpha = 0.75)

# Label the graph
plt.xlabel('Amount of Funding (USD)', fontsize = 20, fontname='Times New Roman')
plt.ylabel('Project Name', fontsize = 20, fontname='Times New Roman')
plt.title('Top 20 Kickstarter Projects by Amount Pledged (in USD)', fontsize= 28)

# Show the graph
plt.tight_layout()
plt.show()
```

- Which parent category has the most funded projects?

To answer this, I used 'usd_pledged' sums to determine which parent category ('category_parent_name') is most funded.

```python
# Groupby the 'category_parent_name' and a sum of 'usd_pledged' 
pledge_category = kickstarter_data.groupby('category_parent_name')['usd_pledged'].sum().sort_values(ascending=False)
pledge_category = pledge_category.iloc[::-1]

# Create the graph - horizontal bar graph
fig, ax = plt.subplots(figsize = (16,9))
plt.barh(pledge_category.index, pledge_category.values, color='y', alpha = 0.75)

# Label the graph
plt.xlabel('Amount of Funding (USD)', fontsize = 20, fontname='Times New Roman')
plt.ylabel('Parent Category', fontsize = 20, fontname='Times New Roman')
plt.title('Top Funded Parent Categories', fontsize = 28)

# Show the graph
plt.tight_layout()
plt.show()
```

- Create a box-plot for the `category_parent_id` column

```python
# Create the subplot
fig, ax = plt.subplots(figsize = (16,9))

# Plot boxplot
plt.boxplot(kickstarter_data['category_parent_id'].dropna()) # Have to drop NA values for boxplotting

# Label the boxplot
plt.xlabel('Boxplot', fontsize = 20, fontname = 'Times New Roman')
plt.ylabel('Category Parent ID', fontsize = 20, fontname = 'Times New Roman')
plt.title('Boxplot of Kickstarter Project Category Parent IDs', fontsize = 28)

# Show the plot
plt.tight_layout()
plt.show()
```

- Create a user-defined function that allows a user to input a category and returns a visualization of your choice.

```python

def customized_visualization(category):
    # Sample data for different categories
    if category == 'a':
        # Create the subplot
        fig, ax = plt.subplots(figsize = (16,9)) 

        # Get data
        data = kickstarter_data['category_name']

        # Plot histogram
        plt.hist(data, bins = 100, label = 'Project Categories', color = 'b', alpha = 0.75) 

        # Label the histogram
        plt.xlabel('Category Name', fontsize = 20, fontname = 'Times New Roman')
        plt.ylabel('Frequency', fontname = 'Times New Roman', fontsize = 20)
        plt.title('Distribution of Kickstarter Categories', fontsize = 28)
        plt.legend(loc = 'upper right', fontsize = 18).get_frame().set_color('#ced6e0')
        plt.xticks(rotation = 80)

        # Show the plot
        plt.tight_layout()
        plt.show()
    
    if category == 'b':
        # Create the subplot
        fig, ax = plt.subplots(figsize=(16, 9))

        # Get data
        data = kickstarter_data['category_name']

        # We must count the frequency of each category as there is no built in frequency function as in histogram
        category_counts = data.value_counts()

        # Plot the bar graph
        plt.bar(category_counts.index, category_counts.values, color='g', alpha=0.75)

        # Label the bar graph
        plt.xlabel('Category Name', fontsize=20, fontname='Times New Roman')
        plt.ylabel('Frequency', fontname='Times New Roman', fontsize=20)
        plt.title('Distribution of Kickstarter Categories', fontsize=28)
        plt.xticks(rotation=80)

        # Show the plot
        plt.tight_layout()
        plt.show()

    if category == 'c':
        # Create the subplot
        fig, ax = plt.subplots(figsize=(50,40))

        # Get data
        data = kickstarter_data['category_name']

        # Calculate the frequencies manually
        category_counts = data.value_counts()
        category_counts = category_counts.iloc[::-1]

        # Plot the bar graph
        plt.barh(category_counts.index, category_counts.values, color='b', alpha=0.75)

        # Label the bar graph
        plt.xlabel('Frequency', fontsize=30, fontname='Times New Roman')
        plt.ylabel('Category Name', fontname='Times New Roman', fontsize=20)
        plt.title('Distribution of Kickstarter Categories', fontsize=28)
        plt.xticks(rotation=0)

        # Show the plot
        plt.tight_layout()
        plt.show()
    

user_category = input("Enter 'a' for histogram, 'b' for vertical bar graph, 'c' or horizontal bar graph")
customized_visualization(user_category)

```

- Creates a dataframe using pandas

- Replace the space with an underscore (_) in all column headers
- Output:

```python
def clean_dataframe(filepath: str):

    df = pd.read_csv(filepath)

    df.columns = df.columns.str.lower().str.replace(' ', '_')

    return df

filepath = '/Users/alexanderbriody/Desktop/Coding Temple/Data-Analytics-Projects/Week of July 30/Day_2/SNAP_history_1969_2019.csv'
data = clean_dataframe(filepath)
data
```
